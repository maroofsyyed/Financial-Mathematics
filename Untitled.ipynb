```python
# create an empty dataframe to store the VIF for each variable
vif = pd.DataFrame()

# calculate VIF using list comprehension 
# use for loop to access each variable 
# calculate VIF for each variable and create a column 'VIF_Factor' to store the values 
vif["VIF_Factor"] = [variance_inflation_factor(df_numeric_features_vif_reduced.values, i) for i in range(df_numeric_features_vif_reduced.shape[1])]

# create a column of variable names
vif["Features"] = df_numeric_features_vif_reduced.columns

# sort the dataframe based on the values of VIF_Factor in descending order
# 'ascending = False' sorts the data in descending order
# 'reset_index' resets the index of the dataframe
# 'drop = True' drops the previous index
vif.sort_values('VIF_Factor', ascending = False).reset_index(drop = True)
```

```python
# for each numeric variable, calculate VIF and save it in a dataframe 'vif'

# use for loop to iterate the VIF function 
for ind in range(len(df_numeric_features_vif_reduced.columns)):
    
    # create an empty dataframe
    vif = pd.DataFrame()

    # calculate VIF using list comprehension
    # use for loop to access each variable 
    # calculate VIF for each variable and create a column 'VIF_Factor' to store the values 
    vif["VIF_Factor"] = [variance_inflation_factor(df_numeric_features_vif_reduced.values, i) for i in range(df_numeric_features_vif_reduced.shape[1])]

    # create a column of variable names
    vif["Features"] = df_numeric_features_vif_reduced.columns

    # filter the variables with VIF greater than 10 and store it in a dataframe 'multi' 
    # one can choose the threshold other than 10 (it depends on the business requirements)
    multi = vif[vif['VIF_Factor'] > 10]
    
    # if dataframe 'multi' is not empty, then sort the dataframe by VIF values
    # if dataframe 'multi' is empty (i.e. all VIF <= 10), then print the dataframe 'vif' and break the for loop using 'break' 
    # 'by' sorts the data using given variable(s)
    # 'ascending = False' sorts the data in descending order
    if(multi.empty == False):
        df_sorted = multi.sort_values(by = 'VIF_Factor', ascending = False)
    else:
        print(vif)
        break
    
    # use if-else to drop the variable with the highest VIF
    # if  dataframe 'df_sorted' is not empty, then drop the first entry in the column 'Features' from the numeric variables
    # select the variable using 'iloc[]'
    # 'axis=1' drops the corresponding column
    #  else print the final dataframe 'vif' with all values after removal of variables with VIF less than 10  
    if (df_sorted.empty == False):
        df_numeric_features_vif_reduced = df_numeric_features_vif_reduced.drop(df_sorted.Features.iloc[0], axis=1)
    else:
        print(vif)
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.compat import lzip
# Names for output
name = ['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value']

# Perform the Breusch-Pagan test using model's residuals and exogenous (independent) variables
test = het_breuschpagan(
    MLR_model_with_significant_var.resid, 
    MLR_model_with_significant_var.model.exog
)

# Print results: lzip combines output names with the test result values for readability
print(lzip(name, test))

# Usually, you care about the f p-value (test[3])
# Interpretation guidance:
if test[3] < 0.05:
    print("Interpretation: The p-value is less than 0.05, so we reject the null hypothesis of homoskedasticity. There is evidence of heteroskedasticity (the variance of residuals is not constant).")
else:
    print("Interpretation: The p-value is greater than 0.05, so we fail to reject the null hypothesis. There is no evidence of heteroskedasticity (variance appears constant).")
```

```python
from scipy.stats import shapiro

# Perform Shapiro-Wilk test on the residuals
stat, p_value = shapiro(MLR_model_with_significant_var.resid)

# Print the test statistic and p-value
print('Shapiro-Wilk Test statistic:', stat)
print('P-Value:', p_value)

# Interpretation guidance:
if p_value < 0.05:
    print("Interpretation: The p-value is less than 0.05. We reject the null hypothesis:")
    print("Conclusion: Residuals are NOT normally distributed.")
else:
    print("Interpretation: The p-value is greater than 0.05. We do not reject the null hypothesis:")
    print("Conclusion: There is no evidence against normality in residuals; they appear normally distributed.")
```

```python
# calculate adjusted R-Squared on train dataset using the formula
# 'n' denotes total number of observations
n = len(X_train_signi_var)

# 'k' denotes number of predictors (including the intercept)
k = len(X_train_signi_var.columns)

# calculate adjusted R-squared
# 'r_sq_mlr' represents the R-Squared value
adj_r_sq = 1 - (((1 - r_sq_mlr) * (n - 1)) / (n - k - 1))

# print the value
adj_r_sq
```

```python
from scipy import stats

# k = number of beta coefficients (including the intercept)
k = len(X_train_signi_var.columns)

# n = total number of observations
n = len(X_train_signi_var)

# r_sq_mlr is your model's R-squared value
f_value = (r_sq_mlr / (k-1)) / ((1-r_sq_mlr)/(n-k))
print("F-statistic value:", f_value)
```

```python
# Degrees of freedom for F statistic
dfn = k - 1  # numerator DOF
dfd = n - k  # denominator DOF

# Calculate F p-value using the survival function (sf)
p_val = stats.f.sf(f_value, dfn, dfd)

print("p-value for the F-statistic:", p_val)

# Interpretation
if p_val < 0.05:
    print("Interpretation: The p-value is less than 0.05. Reject the null hypothesis.")
    print("Conclusion: The overall regression model is statistically significant; at least one beta coefficient is not zero.")
else:
    print("Interpretation: The p-value is greater than 0.05. Fail to reject the null hypothesis.")
    print("Conclusion: The regression model is NOT statistically significant overall.")
```

```python
# define a function to calculate MAPE
# pass the actual and predicted values as input to the function
# return the calculated MAPE 
def mape(actual, predicted):
    return (np.mean(np.abs((actual - predicted) / actual)) * 100)
```
